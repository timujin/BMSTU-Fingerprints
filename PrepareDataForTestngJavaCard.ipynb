```python
pairs_of_fingers = []

for filename in ["mindtct/shelved_pairs_of_fingers/finger_{}.shelve".format(x) for x in range(100)]:
    db = shelve.open(filename)
    pairs_of_fingers.append(db['finger'])
```

```python
from FingerIntoBytes import fingerIntoBytes
pairs_of_finger_arrays = []
for finger1, finger2 in pairs_of_fingers:
    finger1 = [x-128 if x != 256 else 127 for x in fingerIntoBytes(finger1)]
    finger2 = [x-128 if x != 256 else 127 for x in fingerIntoBytes(finger2)]
    
    pairs_of_finger_arrays.append((finger1,finger2))
```

```python
def finger_arrays_into_java_code(farrays, pair):
    def one(x):
        return "new byte[]{{ {} }},".format(','.join([str(y) for y in x]))
    return '\n'.join([one(x[pair]) for x in farrays])
```

```python
print(finger_arrays_into_java_code(pairs_of_finger_arrays[:30], 1))
```
